```python
from matplotlib import pyplot as plt

data_list = ['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY']
#print(data_list)

for column in data_list:
    #•計算任意欄位的平均數及標準差
    print('column [', column ,'] - mean: ', app_train[column].mean())
    print('column [', column ,'] -  std:', app_train[column].std())

    #•畫出任意欄位的直方圖
    plt.hist(app_train['AMT_INCOME_TOTAL'][:500],color='blue',alpha=0.3, label=column)
    plt.legend(loc='best')
    plt.show()
    print('>>> Next:')
```
